# House Price Analysis

## Objective
Analyze property pricing trends and factors affecting price.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/house_prices.csv")

df.head()
```

```python
df.describe()
```

```python
print(df["Price"].mean())
```

```python
location_price = (
    df.groupby("Location")
      ["Price"]
      .mean()
      .sort_values(
          ascending=False
      )
)

location_price
```

```python
sns.scatterplot(
    x="Area",
    y="Price",
    data=df
)

plt.show()
```

```python
sns.boxplot(
    x="Property_Type",
    y="Price",
    data=df
)

plt.show()
```

```python
sns.heatmap(
    df.corr(numeric_only=True),
    annot=True
)

plt.show()
```

## Business Insights

1. Area strongly impacts price.
2. Premium locations command higher prices.
3. Villas show highest pricing.
4. Property age impacts valuation.
5. Housing market trends identified.
