# Supermarket Sales Analysis

## Objective
Analyze sales patterns, customer behavior, product performance, and generate business insights.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/supermarket_sales.csv")

df.head()
```

```python
total_sales = df["Total"].sum()

print("Total Revenue:", total_sales)
```

```python
product_sales = df.groupby("Product_Line")["Total"].sum().sort_values(ascending=False)

product_sales
```

```python
product_sales.plot(kind="bar")
plt.title("Revenue by Product Line")
plt.show()
```

```python
city_sales = df.groupby("City")["Total"].sum()

city_sales.plot(kind="pie", autopct="%1.1f%%")
plt.show()
```

```python
sns.countplot(x="Payment", data=df)
plt.show()
```
