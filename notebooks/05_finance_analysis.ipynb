# Finance Data Analysis

## Objective
Analyze stock performance and market trends.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/finance_data.csv")

df.head()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df["Daily_Return"] = (
    df["Close"]
      .pct_change()
      * 100
)
```

```python
plt.plot(
    df["Date"],
    df["Close"]
)

plt.title("Stock Trend")
plt.show()
```

```python
sns.histplot(
    df["Daily_Return"]
)

plt.show()
```

```python
plt.plot(
    df["Date"],
    df["Volume"]
)

plt.title("Trading Volume")
plt.show()
```

```python
print(
    df["Daily_Return"].std()
)
```

```python
df.mean(numeric_only=True)
```

```python
df.corr(numeric_only=True)
```
