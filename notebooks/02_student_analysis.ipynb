# Student Performance Analysis

## Objective
Analyze academic performance and attendance impact.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/student_performance.csv")

df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df["Average"] = (
    df["Math"] +
    df["Science"] +
    df["English"]
) / 3
```

```python
df["Result"] = np.where(
    df["Average"] >= 40,
    "Pass",
    "Fail"
)

df["Result"].value_counts()
```

```python
df[["Math","Science","English"]].mean().plot(kind="bar")
plt.title("Average Subject Scores")
plt.show()
```

```python
sns.histplot(df["Attendance"])
plt.show()
```

```python
df["Result"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.show()
```

```python
df.mean(numeric_only=True)
```

```python
df.corr(numeric_only=True)
```

## Business Insights

1. Science achieved highest average score.
2. Attendance impacts student performance.
3. Most students passed.
