# Weather Data Analysis

## Objective
Analyze temperature, rainfall, and weather trends.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/weather_data.csv")

df.head()
```

```python
df["Date"] = pd.to_datetime(df["Date"])

df["Month"] = df["Date"].dt.month
```

```python
monthly_temp = (
    df.groupby("Month")
      ["Temperature"]
      .mean()
)
```

```python
monthly_temp.plot(marker="o")
plt.title("Monthly Temperature Trend")
plt.show()
```

```python
sns.boxplot(y=df["Rainfall"])
plt.show()
```

```python
sns.heatmap(
    df.corr(numeric_only=True),
    annot=True
)

plt.show()
```

```python
df[
    df["Temperature"]
    >
    df["Temperature"].quantile(0.95)
]
```
